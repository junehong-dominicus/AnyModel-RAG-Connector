#### Chuncking

```python
# If rag_chunker.py is in the same directory or properly installed in python path
from rag_chunker import ContentChunker

# 1. Initialize
chunker = ContentChunker(
    chunk_size=1000,
    chunk_overlap=200,
    model_name="cl100k_base"
)

# 2. Define some long text
long_text = "Your very long blog post content goes here... " * 500

# 3. Split into chunks
chunks = chunker.split_text(long_text)

# 4. Process chunks
for i, chunk in enumerate(chunks):
    print(f"Chunk {i}: {len(chunk)} characters")
```

#### Embedding

```python
from rag_chunker import ContentChunker
from rag_embedder import LocalEmbedder

# 1. Initialize components
chunker = ContentChunker()
embedder = LocalEmbedder(model_name="text-embedding-embeddinggemma-300m-qat")

# 2. Get your text
raw_text = "Your long blog post content goes here..."

# 3. Chunk the text
chunks = chunker.split_text(raw_text)

# 4. Generate embeddings
if chunks:
    vectors = embedder.embed_documents(chunks)
    print(f"Generated {len(vectors)} embeddings.")
```
